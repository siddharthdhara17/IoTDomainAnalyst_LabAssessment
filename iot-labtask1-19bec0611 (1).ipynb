```python
#reading the dataset as a dataframe df
df=pd.read_csv("../input/diabetes-data-set/diabetes.csv")
```

```python
#printing the head of the dataframe
df.head()
```

```python
#printing the tail of the dataframe
df.tail()
```

```python
#Apply conditions in a column and printing the data that satisfies our condition
#Printing the data for BMI
df[(df['BMI'] == 30)]
```

```python
#Printing data that satisfies any of the conditions
df[(df['BloodPressure']>50) | (df['Age']>25)]
```

```python
#Printing data that satisfies any of the conditions
df[(df['BloodPressure']>50) & (df['Age']>25)]
```

```python
#Make any of the column as index
#Setting date as index
df.set_index("Age", inplace = True)
```

```python
df.head()
```

```python
df.head(10)
```

```python
#Plot any three columns in same graph
df.plot(x="Outcome", y=["BMI","Insulin","BloodPressure"], kind="bar")
```

```python
#Plot one columns in a graph
df["BMI"].plot()
```

```python
#plot bar chart for any one column
df.plot(x="Outcome", kind="bar")
```

```python
#plot stacked bar chart for two columns
df.plot(x="Outcome",y=["BMI"], kind="bar",stacked =True)
```

```python
#plot histogram for any one column
df.plot(x="Outcome",kind="hist")
```

```python
#plot stacked histogram for two columns
df.plot(x="Outcome",y=["Pregnancies","Glucose"], kind="hist", stacked=True)
```

```python
#creating a pivot table
pivot=df.pivot_table(index=['Age'], values=['Outcome'],aggfunc='sum')
pivot
```
